```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
```

```python
(train_data,train_labels),(test_data,test_labels) = fashion_mnist.load_data()
```

```python
#change it to human readable format
class_names=["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
len(class_names)
```

```python
#plot an example images
choice=1700
plt.imshow(train_data[choice],cmap=plt.cm.binary)
plt.title(class_names[train_labels[choice]])
```

```python
#building a Multi class classification model
tf.random.set_seed(42)
model1=tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(10,activation="softmax")
])
model1.compile(loss=tf.keras.losses.CategoricalCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(),
               metrics=["accuracy"])

hist = model1.fit(train_data,tf.one_hot(train_labels,depth=10),epochs=10,validation_data=(test_data,tf.one_hot(test_labels,depth=10)))
```

```python
train_data_norm = train_data/255.0
test_data_norm = test_data/255.0
train_data_norm.min(),train_data_norm.max()
```

```python
#create a model with the input of train data norm
tf.random.set_seed(42)
model2=tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(10,activation="softmax")
])
model2.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(),
               metrics=["accuracy"])

hist_norm = model2.fit(train_data_norm,train_labels,epochs=10,validation_data=(test_data_norm,test_labels))
```

```python
#normalized vs not normalized
import pandas as pd
pd.DataFrame(hist_norm.history).plot(title="normalized data")
pd.DataFrame(hist.history).plot(title="Non normalized data")
```

```python
#finding the ideal learning
tf.random.set_seed(42)
model3=tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(10,activation="softmax")
])
model3.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(),
               metrics=["accuracy"])
#create lr callback
lr_scheduler=tf.keras.callbacks.LearningRateScheduler(lambda epoch:1e-3*10**(epoch/20))
hist_norm3 = model3.fit(train_data_norm,train_labels,epochs=40,validation_data=(test_data_norm,test_labels),callbacks=[lr_scheduler])
```

```python
import numpy as np
lrs = 1e-3*(10**(tf.range(40)/20))
plt.semilogx(lrs,hist_norm3.history["loss"])
plt.xlabel("learning rate")
plt.ylabel("loss");
```

```python
#refit the model with new learning rate
tf.random.set_seed(42)
model4=tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(10,activation="softmax")
])
model4.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
               metrics=["accuracy"])

hist_norm4 = model4.fit(train_data_norm,train_labels,epochs=20,validation_data=(test_data_norm,test_labels))
```

```python
#evaluate our multi class model
from sklearn.metrics import confusion_matrix
yprobs = model4.predict(test_data_norm)
yprobs[0]
 
```

```python
ypreds = yprobs.argmax(axis=1)
ypreds[:10]
```

```python
import random
def plot_random_images(model,images,true_labels,classes):
    i=random.randint(0,len(images))
    #pred and target
    target_img = images[i]
    pred_probs = model.predict(target_img.reshape(1,28,28))
    pred_label = classes[pred_probs.argmax()]
    true_label = classes[true_labels[i]]
    
    plt.imshow(target_img,cmap=plt.cm.binary)
    
    if pred_label == true_label:
        color="green"
    else:
        color = "red"
    
    plt.xlabel("pred: {} {:.2f}% (True: {})".format(pred_label, 100*tf.reduce_max(pred_probs), true_label), color=color)
```

```python
plot_random_images(model=model4,images=test_data,true_labels=test_labels
                   ,classes=class_names);
```
